bandpass filter + export pipeline â€” converting raw .set EEGs into separate band-limited versions.

```python
import os
import mne

def filter_and_save(raw, band, freq_range, save_dir, file_name):
                    #downsample_rate=250):
    """Applies bandpass filter, downsamples EEG data, and saves it in .set format."""
    filtered_raw = raw.copy().filter(l_freq=freq_range[0], h_freq=freq_range[1], fir_design='firwin')
    # filtered_raw.resample(sfreq=downsample_rate)  # Downsample to 50 Hz
    save_path = os.path.join(save_dir, f"{file_name}_{band}.set")
    mne.export.export_raw(save_path, filtered_raw, fmt='eeglab', overwrite=True)
    print(f"Saved {band} band: {save_path}")

def process_eeg_files(root_dir, save_root):
    """Recursively processes .set files from subdirectories and saves filtered/downsampled files."""
    bands = {
        "delta": (0.5, 4),
        "theta": (4, 8),
        "alpha": (8, 13),
        "beta": (13, 30),
        "gamma": (30, 45)
    }

    for sub_dir in os.listdir(root_dir):
        sub_path = os.path.join(root_dir, sub_dir, "eeg")  # Navigate to eeg folder
        if os.path.isdir(sub_path):  # Ensure it's a directory
            for file in os.listdir(sub_path):
                if file.endswith(".set"):
                    file_path = os.path.join(sub_path, file)
                    print(f"Processing {file_path}...")
                    raw = mne.io.read_raw_eeglab(file_path, preload=True)
                    file_name = os.path.splitext(file)[0]

                    for band, freq_range in bands.items():
                        save_dir = os.path.join(save_root, band)
                        os.makedirs(save_dir, exist_ok=True)  # Ensure band folder exists
                        filter_and_save(raw, band, freq_range, save_dir, file_name)

# Define paths
data_root = r"C:\Users\fathi\Desktop\Rest_eeg_ds004504-download\derivatives"

#"C:\Users\fathi\Desktop\ds004504-download\derivatives"
save_root = r"C:\Users\fathi\Desktop\Rest_eeg_ds004504-download\derivatives\500hz_bands_final"

#"C:\Users\fathi\Desktop\ds004504-download\derivatives\bands"

process_eeg_files(data_root, save_root)
```

Converts filtered .set files into .npy NumPy arrays for fast loading in Python

```python
import os
import mne
import numpy as np
from mne import set_log_level

# Set MNE log level to WARNING to reduce output
set_log_level("WARNING")

def convert_set_to_numpy(set_file, numpy_dir):
    """Safely loads a .set file, extracts EEG data, and saves it as a NumPy array."""
    try:
        raw = mne.io.read_raw_eeglab(set_file, preload=True)
        eeg_data = raw.get_data().T  # (timepoints, channels)

        # Create filename and save path
        file_name = os.path.splitext(os.path.basename(set_file))[0]
        numpy_save_path = os.path.join(numpy_dir, f"{file_name}.npy")
        np.save(numpy_save_path, eeg_data)

        # You can comment this line if too many files
        print(f"âœ… Saved: {file_name}.npy | Shape: {eeg_data.shape}")

    except Exception as e:
        print(f"âŒ ERROR in file: {set_file}\n   â†ª {str(e)}")

def process_filtered_eeg_files(filtered_root):
    """Processes .set files in band folders and saves NumPy arrays."""
    bands = ["delta", "theta", "alpha", "beta", "gamma"]

    for band in bands:
        band_dir = os.path.join(filtered_root, band)
        numpy_dir = os.path.join(band_dir, "numpy")
        os.makedirs(numpy_dir, exist_ok=True)

        if os.path.isdir(band_dir):
            for file in os.listdir(band_dir):
                if file.endswith(".set"):
                    file_path = os.path.join(band_dir, file)
                    convert_set_to_numpy(file_path, numpy_dir)

# Set the root directory
filtered_root = r"C:\Users\fathi\Desktop\Rest_eeg_ds004504-download\derivatives\500hz_bands_final"
process_filtered_eeg_files(filtered_root)
```
