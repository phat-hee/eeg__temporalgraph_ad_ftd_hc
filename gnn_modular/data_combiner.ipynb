```python
def combine_band_data(band_data_list):
    combined_dataset = []
    
    min_subjects = min(len(band_data) for band_data in band_data_list)
    
    for subject_idx in range(min_subjects):
        subject_bands = []
        for band_data in band_data_list:
            subject_bands.append(band_data[subject_idx])
        combined_dataset.append(tuple(subject_bands))
    
    return combined_dataset
```
